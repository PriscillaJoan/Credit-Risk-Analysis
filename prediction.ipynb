### Meaning of Terms

- **SK_ID_CURR** - Loan ID for the current loan application.

- **Target** - Indicates whether or not the client has made late loan repayments. (1 = Has made a late loan payment, 0 = Has not made a late loan payment)

- **Name contract type** - Identification if the loan is cash or revolving.
  - A **cash loan** provides a set amount of funds that the borrower receives in a lump sum and repays over a fixed period with regular installments (often monthly).
  - A **revolving loan**, or revolving credit, is a line of credit with a limit that can be borrowed from multiple times, repaid, and borrowed from again, as long as the credit line remains open and the borrower stays within their credit limit.

- **Flag own realty** - Flag if the client owns a house or flat.

- **Amt credit** - The exact figure the borrower receives and is responsible for repaying, along with any interest and fees accrued over time.

- **Loan annuity** - Payments to be made in each installment until the loan is fully paid off.

- **Amt goods price** - For consumer loans, it is the price of the goods for which the loan is given.

- **Name type suite** - Indicates who was accompanying the client when applying for the loan.

- **Name housing type** - The housing situation of the client (e.g., renting, living with parents).

- **Region population relative** - Normalized population of the region where the client lives (higher number means the client lives in a more populated region).

- **Days ID publish** - The number of days before the application that the client changed the identity document used for the loan application.

- **Flag mobil** - Indicates if the client provided a mobile phone (1 = Yes, 0 = No).

- **Flag emp phone** - Indicates if the client provided a work phone (1 = Yes, 0 = No).

- **Flag work phone** - Indicates if the client provided a home phone (1 = Yes, 0 = No).

- **Flag cont mobile** - Indicates if the mobile phone was reachable (1 = Yes, 0 = No).

- **Flag phone** - Indicates if the client provided a home phone (1 = Yes, 0 = No).

- **Flag email** - Indicates if the client provided an email (1 = Yes, 0 = No).

- **Region rating client** - Rating of the region where the client lives (1, 2, or 3).

- **Region rating client w city** - Rating of the region where the client lives, taking the city into account (1, 2, or 3).

- **Weekday appr process start** - The day of the week on which the client applied for the loan.

- **Hour appr process start** - Approximate hour at which the client applied for the loan.

- **Reg region not live region** - Flag if the client's permanent address does not match their contact address (1 = Different, 0 = Same, at the region level).

- **Reg region not work region** - Flag if the client's permanent address does not match their work address (1 = Different, 0 = Same, at the region level).

- **Live region not work region** - Flag if the client's contact address does not match their work address (1 = Different, 0 = Same, at the region level).

- **Reg city not live city** - Flag if the client's permanent address does not match their contact address (1 = Different, 0 = Same, at the city level).

- **Reg city not work city** - Flag if the client's permanent address does not match their work address (1 = Different, 0 = Same, at the city level).

- **Live city not work city** - Flag if the client's contact address does not match their work address (1 = Different, 0 = Same, at the city level).

- **Organization type** - Type of organization where the client works.
